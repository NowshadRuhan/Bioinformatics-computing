```python
import array 
import random


k = input()
k = int(k)

t   = input()
t = int(t)


dna = []
arr = []
for i in range(t) :
    arr = input()
    dna.append(arr)



print(dna)

i=0
j=0
rand = []
arr1 = []
length = len(dna[1])
print(length)

point=0


while i<t :
    
    point = ((length-k)+1) 
    rand_number_for_motif = random.randrange(0, point)
    arr1 = dna[i][rand_number_for_motif:(rand_number_for_motif+k)]
    
    rand.append(arr1)
    i=i+1


print(rand)

p_A = []
p_T = []
p_G = []
p_C = []

i=0
j=0

a = 0.0
tt = 0.0
g = 0.0
c = 0.0

for  i in range(k) : 
    j=0
    for j in range(t) :
        if(rand[j][i] == 'A') :
            a = a+1
            
        elif(rand[j][i] == 'T') :
            tt = tt+1
            
        elif(rand[j][i] == 'G') :
            g = g+1
            
        elif(rand[j][i] == 'C') :
            c = c+1 
            
        j=j+1 
    
    a = a/t
    p_A.append(a)
    a = 0.0
    tt = tt/t
    p_T.append(tt)
    tt = 0.0
    g = g/t
    p_G.append(g)
    g=0.0
    c = c/t
    p_C.append(c)
    c = 0.0
    
    
print("Profile Matrix :")

print("A :")
print(p_A)

print("T :")
print(p_T)

print("G :")
print(p_G)

print("C :")
print(p_C)


#input:

#3
#4
        
#CAAATGCA
#TTCATGAC
#GACAGGGC
#TCACTTGT
```
