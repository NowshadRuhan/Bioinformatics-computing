```python
#SKEW DIAGRAM Code for all

#run it on Jupyter Nootbook

import matplotlib.pyplot as plt
d   = input()
z = len(d)
x = [100]
y = [100]
m = 0
y[0] = 0
x[0] = 0
i=0
for i in range (z):
  
    if(d[i] == 'C'):
        m = m-1
        y.append(m)
    
    elif(d[i] == 'G'):
        m = m+1
        y.append(m)
    
    else:
        y.append(m)

    x.append(i+1)

plt.plot(x, y, color='green', linestyle='dashed', linewidth = 5, 
         marker='o', markerfacecolor='blue', markersize=12)

plt.
plt.xlabel('X Label ') 
plt.ylabel('Y Label') 
plt.title('SKEW DIAGRAM') 
plt.show() 


#input
#ATCCCGGA


```
